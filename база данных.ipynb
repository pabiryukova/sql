### 1. С помощью запроса выберите фильмы ранее 2018 года


```python
question1 = """
SELECT title, premiered
FROM titles 
WHERE premiered < 2018
ORDER BY premiered DESC
"""
cur.execute(question1)
cur.fetchall()
```

### 2. 50 самых популярных фильмов (по кол-ву оценок)

```python
question2 = """
SELECT title, votes
FROM titles 
    JOIN rating ON  titles.title_id = rating.title_id
GROUP BY title
ORDER BY votes DESC
LIMIT 50
"""

cur.execute(question2)
cur.fetchall()
```

### 3.10 самых популярных (по кол-ву оценок и с рейтингом >= 7) комедий 2019 года

```python
question3 = """
SELECT  title, votes, rating
FROM titles 
    JOIN rating ON  titles.title_id = rating.title_id
    JOIN film_genres ON rating.title_id = film_genres.title_id
    JOIN genre_types ON film_genres.genre_id = genre_types.id
   WHERE genre_name = 'Comedy' AND premiered = 2019 AND rating>= 7
ORDER BY votes DESC
LIMIT 10
"""

cur.execute(question3)
cur.fetchall()
```

### 4. С помощью запроса выберите год выхода и название фильмов с кем-то из ваших любимых актеров

```python
question4 = """
SELECT title, premiered
FROM titles 
    JOIN crew ON titles.title_id = crew.title_id
    JOIN people ON crew.person_id = people.person_id
WHERE name = "Leonardo DiCaprio" AND premiered IS NOT NULL
ORDER BY premiered DESC
"""
cur.execute(question4)
cur.fetchall()
```

### 5. Выберите жанр и посчитайте кол-во фильмов по годам


```python
question5 = '''
SELECT premiered, COUNT(genre_name) as comedy
FROM titles 
    JOIN film_genres ON titles.title_id = film_genres.title_id
    JOIN genre_types ON film_genres.genre_id = genre_types.id
   WHERE genre_name = 'Comedy' AND premiered IS NOT NULL
GROUP BY  premiered
ORDER BY premiered DESC

'''

cur.execute(question5)
cur.fetchall()
```

### 7. распределение оценок для комедий и документальных фильмов

```python
question7 = '''
SELECT  ROUND(rating, -1) as Оценка,
    sum(case when genre_name = 'Comedy' then 1 else 0 end) as Комедии,
    sum(case when genre_name = 'Documentary' then 1 else 0 end) as Документалки
FROM titles 
    JOIN rating ON  titles.title_id = rating.title_id
    JOIN film_genres ON titles.title_id = film_genres.title_id
    JOIN genre_types ON film_genres.genre_id = genre_types.id

    GROUP BY Оценка
ORDER BY Оценка DESC

'''

cur.execute(question7)
cur.fetchall()
```
